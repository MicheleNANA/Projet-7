### 2.1 Table Disponibilité alimentaire 2017

```python
# Importation de la table Disponibilité alimentaire de 2017

dispo = pd.read_csv('DisponibiliteAlimentaire_2017.csv')
dispo.head(3)
```

```python
# Informations de la table

dispo.info()
```

```python
dispo.head(3).T
```

```python
# Valeurs uniques de la colonne "Produit"

dispo['Produit'].unique()
```

```python
# Filtre des données de "viande de volailles"

viande_volaille = dispo.loc[dispo['Produit']== 'Viande de Volailles']
viande_volaille.tail(3)
```

```python
# Informations du df filtré

viande_volaille.info()
```

```python
# Colonnes du df filtré

viande_volaille.columns
```

```python
# Valeurs uniques dans chaque colonne

for i in viande_volaille.columns:
  print(i, viande_volaille[i].unique())
```

```python
# Variables à sélectionner

volaille_select = viande_volaille.loc[:,['Zone','Élément','Produit', 'Unité', 'Valeur']]
volaille_select.head()
```

```python
# Informations du nouveau df

volaille_select.info()
```

```python
# Faire pivoter la colonne "Élément"

volaille_pivot = volaille_select.pivot_table(index='Zone', columns='Élément', values = 'Valeur')
volaille_pivot.head(3)
```

```python
# Valeurs nulles par variables (Colonne)

volaille_pivot.isnull().sum()
```

```python
# Remplacer les valeurs manquantes des colonnes par leurs moyennes

volaille_pivot = volaille_pivot.fillna(volaille_pivot.mean())
volaille_pivot.head(3)
```

P4 : Production + Importations - Exportations + Variation de stock = Disponibité intérieure = Semences + Pertes + Nourriture + Aliments pour animaux + Traitement + Autres sutilisations

Taux de dépendance aux importations (TDI) = (Importation ÷ Disponibilité intérieure) x 100

Taux d'auto-suffisance (TAS) = (Production ÷ Disponibilité intérieure) x 100


```python
# Calcul du Taux de dépendance aux importations (TDI)

volaille_pivot['TDI(%)'] = (volaille_pivot['Importations - Quantité']/volaille_pivot['Disponibilité intérieure'])*100
volaille_pivot.head(3)
```

```python
# Calcul du Taux d'auto-suffisance (TAS)

volaille_pivot['TAS(%)'] = (volaille_pivot['Production']/volaille_pivot['Disponibilité intérieure'])*100
volaille_pivot.head(3)
```

```python
# Informations du df

volaille_pivot.info()
```

```python
# Variables à sélectionner

volaille = volaille_pivot.loc[:,['Disponibilité alimentaire en quantité (kg/personne/an)','Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité de protéines en quantité (g/personne/jour)', 'TDI(%)', 'TAS(%)']]
volaille.head(3)
```

### 3.2 Les jointures

```python
# 1. Distance = Jointure Distance_Zone

distance = distancefrance.merge(zone, how = 'left', on = 'iso_d')
columns_to_drop = ['iso_d']
distance.drop(columns=columns_to_drop, inplace=True)
distance.head()
```

```python
# Informations

distance.info()
```

```python
# 2. Faodistance = jointure Faostat_Distance

faodistance = fao.merge(distance, how = 'left', on = 'Zone')
faodistance.head()
```

```python
# Informations

faodistance.info()
```

```python
# 3. Avec pop_croissance_demo

faodistpop = faodistance.merge(pop, how = 'left', on = 'Zone')
faodistpop.head()
```

```python
# Informations

faodistpop.info()
```

```python
# 4. Avec volaille

data = faodistpop.merge(volaille, how = 'right', on ='Zone')
data.head()
```

```python
# Informations

data.info()
```

### 3.3 Filtrage et nettoyage du dataframe créé

```python
# Renommer certaines colonnes

data.rename(columns = {'dist': 'distance_France(km)' }, inplace = True)
data.rename(columns = {'2017': 'population_2017' }, inplace = True)
data.rename(columns = {'PIB par habitant, ($ PPA internationaux constants de 2017)': 'PIB' }, inplace = True)
data.rename(columns = {'Stabilité politique et absence de violence/terrorisme (indice)': 'stabilite_politique' }, inplace = True)
data.head()
```

```python
# Filtre de la France

france = data[data['Zone'] == "France"]
france
```

```python
# Exclusion de la France des données!

data = data.drop(data.loc[data['Zone'] == "France"].index)
```

```python
# Sélection des variables pour la suite

data = data[['Zone', 'PIB', 'TDI(%)', 'TAS(%)', 'croissance_demographique(%)', 'distance_France(km)', 'population_2017', 'stabilite_politique', 'Disponibilité alimentaire en quantité (kg/personne/an)', 'Disponibilité alimentaire (Kcal/personne/jour)', 'Disponibilité de protéines en quantité (g/personne/jour)']]
data.head(3)
```

```python
# Les valeurs manquantes

data.isnull().sum()
```

```python
# Afficher les lignes de PIB qui ont des valeurs manquantes

data[data.PIB.isna()]
```

```python
# Définir la variable 'Zone' comme index

data.set_index('Zone', inplace=True)
data.head(3)
```

```python
# Remplacer les valeurs manquantes par les moyennes

data = data.fillna(data.mean())
data.head(3)
```

```python
# Filtrer les lignes où l'index contient le nom "Chine"

chine_data = data[data.index.str.contains('Chine', case=False, na=False)]

# Afficher les résultats

chine_data
```

```python
# Supprimer les autres Chine

index_to_remove = ['Chine - RAS de Hong-Kong', 'Chine - RAS de Macao', 'Chine, Taiwan Province de']

data = data.drop(index_to_remove)
data.shape
```
